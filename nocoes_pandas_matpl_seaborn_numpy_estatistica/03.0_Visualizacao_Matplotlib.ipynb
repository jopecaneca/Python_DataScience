```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Graficos de linha

```python
np.random.seed(7)
# vetor com números inteiros aleatótrios entre 1 a 1500, com 15 amostras

y = np.random.randint(low=1, high=1750, size=15)
```

```python
y
```

```python
plt.plot(y)
```

```python
# Trabalhando no gráfico anterior para melhora-lo

# inserir a primeira linha no plot
plt.plot(y,color = '#749187', marker = 'o', ms = 5, mec ='k',
        markerfacecolor = 'w', ls = '-.')

# Inserir a segunda linha no plot
plt.plot(y*2, marker = '+', color = 'k', ms = 5)

# rótulos
plt.xlabel("Eixo X", color = 'red', size =12)
plt.ylabel("Eixo Y")
plt.title("Grafico para praticar Storytelling com Dados - Caneca", loc= 'left')

# gridlines

plt.grid(axis='both', color = 'gray',
        linestyle = 'solid', linewidth = 1,alpha = 0.8)

plt.show()
```

```python
#subplots
np.random.seed(5)
x = np.arange(1, 11)
y1 = np.random.randint(1, 20, 10)
y2 = np.random.randint(10, 500, 10)
y3 = np.random.randint(150, 450, 10)
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Figure')
ax[0].plot(x, y1, color='black')
ax[1].plot(x, y2, color='red')
ax[2].plot(x, y3, color='green')

for i in range(3) :
    ax[i].set(title = f'Subplot {i+1}', xlabel ='Eixo X', ylabel ='Eixo Y')
```

#### Scatter plot

```python
x = np.random.normal(10, 0.5, 100)
y = np.random.uniform(0, 20, 100)
```

```python
fig = plt.figure()
ax = plt.axes()
ax.scatter(x , y, marker= 'o', color='black', label = 'data 1', alpha=0.5)
ax.scatter(x*0.5, y*0.5, marker= 'x', color='green', label = 'data 2', alpha=0.7)
ax.legend();
```

```python
dados = px.data.gapminder().query('country == "Angola"').set_index('year')
```

```python
dados.tail()
```

```python
dados = dados.reset_index()
```

```python
dados
```

```python
# Renomar as colunas usando um dicionario
dados.rename(
    columns={
        'year' : 'Ano',
        'country' : 'Pais',
        'continent' : 'Continente',
        'lifeExp' : 'Expetativa_Vida',
        'pop' :'Populacao',
        'gdpPercap' : 'Renda_Percapita'
    }, inplace=True
)

# Verificar
dados.columns
```

```python
dados
```

```python
plt.plot(dados['Ano'], dados['Renda_Percapita'])
plt.title('PIB per capita de Angola')
plt.ylabel('Per Capita')
plt.xlabel('Ano');
```

```python
titulo = 'Relacao entre a expectativa de vida e a renda per capita em Angola'

plt.figure(figsize= (8, 4))

plt.scatter(dados['Expetativa_Vida'], dados['Renda_Percapita'] )
plt.title(titulo, loc='left')
plt.ylabel('Per Capita')
plt.xlabel('Expectativa de Vida')
plt.show();
```

```python
plt.bar(x= dados.Ano,
        height= dados['Populacao'],
        color= 'black'
        )
plt.title('Populacao Angolana');
```
